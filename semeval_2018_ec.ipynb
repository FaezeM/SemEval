Should we ommit the @s or no?

```python
all_mentions = set()
for tweet in df['Tweet']:
    mentions = re.findall(r'@(\w+)', str(tweet))  # \w+ = username (letters/numbers/_)
    all_mentions.update(mentions)

print(f"Mentions used are ({len(all_mentions)} total):")
all_mentions
```

```python
# Extract all mentions and count frequencies
all_mentions = re.findall(r'@(\w+)', ' '.join(df['Tweet'].astype(str)))
mention_counts = Counter(all_mentions)

# Filter mentions with count > 1
repeated_mentions = {mention: count for mention, count in mention_counts.items() if count > 3}

# Sort by frequency (descending)
sorted_repeats = sorted(repeated_mentions.items(), key=lambda x: x[1], reverse=True)

# Results
print(f"Mentions used more than once ({len(sorted_repeats)} total):")
for mention, count in sorted_repeats:
    print(f"@{mention}: {count}")
```

Clean mentions except the repeated ones

```python
def filter_mentions(text, important_dict):
    return re.sub(r'@\w+', lambda m: m.group(0) if m.group(0)[1:] in important_dict else '', text)
```

```python
df['Tweet'] = df['Tweet'].apply(lambda x: filter_mentions(x, repeated_mentions))
```

Lowercase but ALL CAPS

```python
def find_all_caps(txt):
    return re.sub(r'\b(?![A-Z]{2,}\b)([A-Za-z]+)\b', lambda m: m.group(0).lower(), txt)
```

```python
df['Tweet'] = df['Tweet'].apply(find_all_caps)
```

Clean the URLs

Clean the hashtag symbols

```python
df['Tweet'] = df['Tweet'].str.replace(r'#(\w+)', r'\1', regex=True)
```
